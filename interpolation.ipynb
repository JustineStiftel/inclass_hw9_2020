```python
from scipy.interpolate import interp1d
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,num=11,endpoint=True)
```

```python
y=np.cos(-x**2/9.)
```

```python
f=interp1d(x,y)
```

```python
f2=interp1d(x,y,kind='cubic')
```

```python
xnew=np.linspace(0,10,num=41,endpoint=True)
ynew=np.cos(-xnew**2/9.)
xref=np.linspace(0,10,num=41,endpoint=True)
yref=np.cos(-xref**2/9.)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y,'o',xnew,f(xnew),'-',xnew,f2(xnew),'--',xref,ynew,'k.')
plt.legend(['data','linear','cubic'],loc=2)
```

```python
plt.show()
```
